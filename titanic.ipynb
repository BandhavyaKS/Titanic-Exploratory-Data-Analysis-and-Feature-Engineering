```python
## fill the null values, we useboxplot( it is used to find outliers)
## here lets use 0 or mean value
# data ->numerical
#      ->categorical

age_mean=df['age'].mean()
age_mean
```

```python
fare_mean=df['fare'].mean()
fare_mean
```

```python
## cabin has more null values

# embarked, is categorical ( ) , use mode( highest no of occurance,repeated values)

df['embarked'].unique() 
```

```python
embarked_mode=df['embarked'].mode()[0] # 0 bcz first most repeated value
embarked_mode
```

```python
## after finding the null values , we have to fill the null values
df['age']=df['age'].fillna(age_mean)
df['fare']=df['fare'].fillna(fare_mean)
df['embarked']=df['embarked'].fillna(embarked_mode)
```

```python
# drop the name column
df=df.drop(columns='name')
df
```

```python
## while creating the table, if the table creates extra none column , so give n=1
df[['title','F_name']]=df['F_name'].str.split('.',expand=True,n=1)
df
```

```python
## to remove the spaces
df['title']=df['title'].str.strip()
df['F_name']=df['F_name'].str.strip()
df['title']
df['F_name']
```

```python
## round of the age values
df['age']=df['age'].round(1)
df
```

```python
## to capitalize each column name
df.columns=df.columns.str.capitalize()
df
```

```python
## sibling , spouse, parent and chid are from same fam,lets use count

df['Fam_count']=df['Sibsp']+df['Parch']
df
```

```python
## round the number to single digit
df['Fare']=df['Fare'].round(1)

df
```

```python
## check the embarked column

df['Embarked'].unique()
## we need to write the full name
# S=Southampton
# C=Cherbourge
# Q=Queenstown
## to replace the values let us use the dict
```

```python
d={'S':'Southampton','C':'Cherbourge','Q':'Queenstown'}
df['Embarked']=df['Embarked'].replace(d)
df
```

```python
## sort_values is to arrange the records, on what we pass
## by default there is ascen=nding
df.sort_values(by=['Ticket'],ascending=True)

df=df.sort_values(by=['Ticket','Age','Fare'],ascending=[True,True,True])
df
```

```python
## arrange then order vise
df=df.reset_index(drop=True)
df

```

```python
## some passengers are sharing same ticket, to check the couont
## reset_index Creates the table format
df['Ticket'].value_counts().reset_index() ## it give s he frequency 
```

```python
## i want to change the count to , no_of passengers
## it creates separate table (Dataframe) forr ticket
Ticket_pass=df['Ticket'].value_counts().reset_index(name='No_of_Passengers')
Ticket_pass
```

```python
df
```

```python
## merge the Ticet _pass and df
# merge func:--------is used to merge the two table
# we have to pass using pandas
# merge has parameters:
# ----->>>>>on=,left_on=,right_on=,hoe=
# ----------on is the common column
# ----------how='inner'(inner will perform inner join condition)( merging with known records)
# ---------------'outer'(performs outer join)
# ---------------'left'()
# ---------------'right'()
# ----------------above are same like sql

# if the colmn name is diff in both table(same data, diff col names)
#     then give both
#             left table=df
#             right table= ticket_pass

df=pd.merge(df,Ticket_pass,on='Ticket',how='inner')
df
```

```python
## companion type
## for ticket column first 3 are same
# to see who are travelling together total count, even tough family count 3 , it means friends......
# dividde among 
df['Fare']=(df['Fare']/df['No_of_Passengers']).round(1)
df

```

```python
## paramter=row-----------( in pandas the data are in structure of row and column ( it checks data in row by row))
## F= STORES FAM _COUNT
    #no of passengers internally searches like key=no of passengers, its val i.e 3 as values
## P STORE NO OF PASSENGERS
# we mention row,, 
## if no of passengers is 0 means solo
## if fam count is zero and no of passsengers are greater than 1 means returf friends
# else fam count present it is family



def comp_type(row):
    f=row['Fam_count']
    p=row['No_of_Passengers']
    if p==1:
        return 'Solo'
    elif f==0 and p>1:
        return 'Friends'
    else:
        return 'Family'
```

```python
## APPLY
        ##it is used for complex condition
        
# it is a looping func
# in python we use for loop here we use apply




df['Companion_type']=df.apply(comp_type,axis=1)
df
```

```python
## instaead of creating all these , use CUT func
## CUT: built in methods, when we have range of values, only for numerical data
## cut it accepsts bins, 0,2,16 if the range is satisfied
## if the range avbove all, it returns NONE



def Age_category(row):
    if Age >=0 and Age<=2:
        return 'Infant'
    elif Age>2 and Age<=16:
        return 'Child'
    elif Age>16 and Age<=25:
        return 'Teen'
    elif Age>25 and Age<50:
            return 'Adult'
    elif age>50 and Age<=100:
        return 'Senior'
    
    
    
    
        
```

```python
df['Age  Category']=pd.cut(df['Age'],bins=[0,2,16,25,59,100],labels=['Infant','Child','Teen','Adult','Senior'])
```

```python
## According to the title column we will display the social stus of the paseengers
## the countless--Jonkheer( represent the ing or queen in dutch)
## REV--- or a person in temple, ( for priest)
##Mlle- MADAM in france is married woman
##Mme --MADAM in france is unmarried woman
# DON - SPANISH WORD FOR  MALE NOBEL PERSON
## DONA-- SPANISH WORD FOR  WOMAN NOBEL PERSON



df['Title'].unique()
```

```python
Royals=['the Countess','Jonkheer ','Don', 'Dona','Master','Sir']
Common_person=['Miss','Mrs', 'Mr','Ms', 'Mme', 'Mlle']
Militry=[ 'Major','Col''Capt']
Other_professionals=[ 'Dr','Rev']
```

```python
def social_status(row):
    c=row['Title']
    if c in Royals:
        return 'Royals'
    elif c in Common_person:
        return 'Common_person'
    elif c in Militry:
        return 'Militry'
    else:
        return Other_professionals
    
df['Social_status']=df.apply(social_status,axis=1)
    
    
```
